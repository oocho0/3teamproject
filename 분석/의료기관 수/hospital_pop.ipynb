```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rc

rc('font', family='AppleGothic')
matplotlib.rcParams['axes.unicode_minus'] = False

hs_1=pd.read_csv('병원.csv')
hs_2=pd.read_csv('종합병원.csv')
hs_3=pd.read_csv('상급종합병원.csv')
hs_h=pd.read_csv('요양병원.csv')
pop=pd.read_csv('../89개시군구인구/total_pop.csv')
sido=pd.read_csv('../total_pop/sido.csv')
```

```python
pop.set_index('sigungu', inplace=True)
pop.sort_index(inplace=True)
pop
```

```python
sido.set_index('sigungu', inplace=True)
sido.sort_index(inplace=True)
sido
```

```python
hs_1.set_index('sigungu', inplace=True)
hs_1.sort_index(inplace=True)
hs_1

hs_2.set_index('sigungu', inplace=True)
hs_2.sort_index(inplace=True)
hs_2

hs_3.set_index('sigungu', inplace=True)
hs_3.sort_index(inplace=True)
hs_3

hs_h.set_index('sigungu', inplace=True)
hs_h.sort_index(inplace=True)
hs_h
```

```python
hs_t=pd.DataFrame({'hospital_total':(hs_1['2021.4/4']+hs_2['2021.4/4']+hs_3['2021.4/4']
                                     +hs_h['2021.4/4'])})
hs_t['hs_1']=hs_1['2021.4/4']
hs_t['hs_2']=hs_2['2021.4/4']
hs_t['hs_3']=hs_3['2021.4/4']
hs_t['hs_h']=hs_h['2021.4/4']
hs_t['sido']=sido['sido']
hs_t['pop']=pop['2021']
hs_t
```

```python
hs_rate_t=pd.DataFrame({'hospital_total':(hs_1['2021.4/4']+hs_2['2021.4/4']+hs_3['2021.4/4']
                                     +hs_h['2021.4/4'])/pop['2021']*100000})
hs_rate_t['hs_1']=hs_1['2021.4/4']/pop['2021']*100000
hs_rate_t['hs_2']=hs_2['2021.4/4']/pop['2021']*100000
hs_rate_t['hs_3']=hs_3['2021.4/4']/pop['2021']*100000
hs_rate_t['hs_h']=hs_h['2021.4/4']/pop['2021']*100000
hs_rate_t['sido']=sido['sido']
hs_rate_t['pop']=pop['2021']
hs_rate_t
```

```python
import numpy as np

count_t, bin_dividers_t = np.histogram(hs_rate_t['hospital_total'],bins=4)
bin_dividers_t
```

```python
count_1, bin_dividers_1 = np.histogram(hs_rate_t['hs_1'],bins=4)
bin_dividers_1
```

```python
count_2, bin_dividers_2 = np.histogram(hs_rate_t['hs_2'],bins=4)
bin_dividers_2
```

```python
count_3, bin_dividers_3 = np.histogram(hs_rate_t['hs_3'],bins=4)
bin_dividers_3
```

```python
count_h, bin_dividers_h = np.histogram(hs_rate_t['hs_h'],bins=4)
bin_dividers_h
```

```python
bin_names_t=['0~14','14.1~28','28.2~42','42.3~56.6']
bin_names_1=['0~4','4.6~9','9.2~13','13.9~18.6']
bin_names_2=['0~5','5.6~11','11.3~16','16.9~22.6']
bin_names_3=['0~0.5','0.6~1','1.3~1.8','1.9~2.6']
bin_names_h=['0~9','9.1~18','18.2~27','27.4~36.6']

hs_rate_t['hp_bin_t']=pd.cut(x=hs_rate_t['hospital_total'],bins=bin_dividers_t,labels=bin_names_t,include_lowest=True)
hs_rate_t['hp_bin_1']=pd.cut(x=hs_rate_t['hs_1'],bins=bin_dividers_1,labels=bin_names_1,include_lowest=True)
hs_rate_t['hp_bin_2']=pd.cut(x=hs_rate_t['hs_2'],bins=bin_dividers_2,labels=bin_names_2,include_lowest=True)
hs_rate_t['hp_bin_3']=pd.cut(x=hs_rate_t['hs_3'],bins=bin_dividers_3,labels=bin_names_3,include_lowest=True)
hs_rate_t['hp_bin_h']=pd.cut(x=hs_rate_t['hs_h'],bins=bin_dividers_h,labels=bin_names_h,include_lowest=True)
hs_rate_t
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='whitegrid')

sns.scatterplot(x='pop', y='hospital_total', hue='sido',style='sido',s=60,data=hs_t)

plt.title('인구감소지역 전체 병원 수')
plt.xlabel('인구수')
plt.ylabel('1,2,3차,요양 병원 수')
plt.legend(fontsize=6)
plt.savefig('output/hs_t_pop_sido.png', bbox_inches='tight')
plt.show()
```

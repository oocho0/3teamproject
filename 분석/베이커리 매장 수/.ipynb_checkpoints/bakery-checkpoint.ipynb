```python
import pandas as pd

bakery = pd.read_csv('./bakery.csv')
sig_cd = pd.read_csv('../total_pop/sig_cd.csv')
bakery
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='whitegrid')

plt.scatter(bakery['뚜레쥬르'], bakery['파바'], s = 5)

for i in range(len(bakery)): # 행 개수만큼 순회
    row = bakery.iloc[i] # 한 행씩 꺼내기
    name = row['sigungu'] # 이름이 저장된 열
    x = row['뚜레쥬르'] # x좌표가 저장된 열
    y = row['파바'] # y좌표가 저장된 열
    if name in ['충청북도 제천시']:
        plt.text(x-0.5, y-0.5, name, fontsize=6)
    elif name in ['경상북도 안동시']:
        plt.text(x-0.8, y+0.2, name, fontsize=6)
    elif name in ['대구광역시 서구','대구광역시 남구','부산광역시 동구','충청남도 보령시',
                  '부산광역시 서구', '경기도 가평군'] : 
        plt.text(x-0.5, y+0.2, name, fontsize=6) # x 좌표, y좌표, 이름 순서로 input 지정
    elif name in ['경상북도 영천시', '충청남도 논산시'] :
        plt.text(x-0.5, y+0.5, name, fontsize=6)
    elif name in ['경상북도 영주시'] :
        plt.text(x-0.5, y+0.8, name, fontsize=6)
    elif name in ['전라북도 정읍시'] : 
        plt.text(x-0.5, y+1.1, name, fontsize=6)
    else :
        pass

plt.title('인구감소지역 뚜레쥬르-파리바게트 매장 수')
plt.xlabel('뚜레쥬르 매장')
plt.ylabel('파리바게트 매장')
plt.savefig('output/bakery.png', bbox_inches='tight')
plt.show()
```

```python
bakery.set_index('sigungu',inplace=True)
bakery.sort_index(inplace=True)
bakery
```

```python
pop = pd.read_csv('../total_pop/pop_89.csv')
pop.rename(columns={'행정구역(시군구)별':'sigungu'},inplace=True)
pop.set_index('sigungu',inplace=True)
pop.sort_index(inplace=True)
pop
```

```python
sido=pd.read_csv('../total_pop/sido.csv')
sido.set_index('sigungu', inplace=True)
sido.sort_index(inplace=True)
sido
```

```python
bakery['pop']=pop['total_2020']
bakery['sido']=sido['sido']
bakery
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='whitegrid')

sns.scatterplot(x='pop', y='sum', hue='sido',style='sido',s=60,data=bakery)

plt.title('인구감소지역 뚜레쥬르-파리바게트 매장 수')
plt.xlabel('인구수')
plt.ylabel('매장 수')
plt.legend(fontsize=6)
plt.savefig('output/bakery_pop_sido.png', bbox_inches='tight')
plt.show()
```

```python
import numpy as np

count, bin_dividers=np.histogram(bakery['sum'], bins=4)
bin_dividers
```

```python
bin_names=['0~4개', '4.5~8개','9~13개','13.5~18개']

bakery['hp_bin']=pd.cut(x=bakery['sum'], bins=bin_dividers, labels=bin_names, include_lowest=True)
bakery
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='whitegrid')

sns.scatterplot(x='pop', y='sum', hue='hp_bin',style='hp_bin',s=60,data=bakery)

plt.title('인구감소지역 뚜레쥬르-파리바게트 매장 수')
plt.xlabel('인구수')
plt.ylabel('매장 수')
plt.legend(fontsize=6)
plt.savefig('output/bakery_pop_bins.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='AppleGothic')

popbyage=pd.read_csv('popbyage.csv', encoding='cp949')
popbyage
```

```python
popbyage.drop(['시점','성별','계', '80세 이상','85세 이상'], axis=1, inplace=True)
popbyage.set_index('행정구역(시군구)별', inplace=True)
popbyage
```

```python
popbyage=popbyage.T
popbyage
```

```python
popbyage.info()
```

```python
locals=['임실군','순창군','고창군','부안군','보성군','장흥군','강진군','해남군','신안군','고령군','울릉군','하동군','합천군']

for local in locals :
    fig, ax = plt.subplots()
    popbyage.plot(kind='bar',y=local)
    #plt.show()
    plt.savefig('output/'+local+'.png')
    plt.close(fig)
    
```
